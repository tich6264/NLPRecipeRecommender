# Format spaCy v3 Recipe Data

### Import spaCy dependencies

```python
import spacy
from spacy.tokens import DocBin
```

### Read in pre-parsed recipe data from csv

#### Training data takes the following example format:
training_data = [("Tokyo Tower is 333m tall.", [(0, 11, "BUILDING")])]

```python
training_data = []
with open("../intermed_data/recipe_training_data_v3.csv", "r") as f: 
    for line in f:
        line = line.strip() # remove new line character
        line = eval(line)
        training_data.append(line)
print(training_data)
```

### Process recipe data into spaCy v3 format
Format prepped recipe data into spaCy v3 binary format by serializing a DocBin (i.e. collection of Doc objects)
    

```python
nlp = spacy.blank("en")
folder_name = "../"

# the DocBin will store the example documents
db = DocBin()
i = 0
for text, annotations in training_data:
    try:
        doc = nlp(text)
        ents = []
        
        for start, end, label in annotations:
            span = doc.char_span(start, end, label="INGREDIENT")
            ents.append(span)
        doc.ents = ents
        db.add(doc)
       
        if(i % 100000 == 0):
            
            file_name = "../../training/data/recipe_train_"+ str(i) + ".spacy"
            db.to_disk(file_name)
            db = DocBin()
        i = i+1
        
    except: 
        # TODO: if time permits, debug annotation parsing errors
        # Note: some errors arise from the dataset itself - needs some cleaning
        print("failure to insert------------------")
        print("text")
        print(text)
        print("annotations")
        print(annotations)
    

file_name = "../../training/data/recipe_train_"+ str(i) + ".spacy"
db.to_disk(file_name)
```
