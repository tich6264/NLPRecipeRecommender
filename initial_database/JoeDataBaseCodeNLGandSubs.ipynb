```python
def simple_tokenize(text):
  tokens=[''.join(letter for letter in word if letter not in string.punctuation) for word in text.split()]
  tokens = [s.strip() for s in tokens]
  return tokens
```

```python
def get_idf(df, N):
  return np.log(N/df)
```

```python
# place recipe titles into a list and begin
# cleaning process by converting titles to lower-case
recipeTitles = []
cleanRecipeTitles = []

for index, row in df.iterrows():
  recipeTitles.append(row.title.lower())

for title in recipeTitles:
  cleanTitle = simple_tokenize(title)
  cleanRecipeTitles.append(cleanTitle)

cleanRecipeTitles
N = len(cleanRecipeTitles)
print(N)
```

# Calculate Document Frequency for Analysis Threads

```python
all_terms = []

for recipe in cleanRecipeTitles:
  for word in recipe:
    if word != '':
      all_terms.append(word)

all_terms = list(set(all_terms))

DocFrequency = {term: 0 for term in all_terms}

for term in all_terms:
  for title in cleanRecipeTitles:
    if term in title:
      DocFrequency[term] += 1

DocFrequency
```

```python
from collections import Counter
import numpy as np

index = 0

# define dictionary to hold Unigram TF-IDF values
tf_idf = {}

# Using file_dict loaded in Section 1.3.2, read values to Term Frequency for each document
for recipe in cleanRecipeTitles:
  token_length = len(recipe)
  count = Counter(recipe)
  
  for token, number in count.items():
    # calculate term frequency
    tf = number / token_length

    # calculate document frequency
    if token != '':
      df = DocFrequency[token]

    # idf calculation
    idf = get_idf(df, N)

    # tf-idf calculation
    tf_idf[index, token] = tf*idf

  # increment index
  index += 1

for x,y in tf_idf.items():
  print(x, y)
  
```
